```python
import numpy as np
import cv2
from skimage import color, data, restoration, util
```

```python
img = cv2.imread('Images/you.jpg')
img = cv2.resize(img,(700, 920))
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Unction modelling
h=np.eye(12)/12
unction=cv2.filter2D(img,-1,h)
cv2.imshow('unction image', unction)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Restoration by Lucy-Richardson
deconvolved = restoration.richardson_lucy(unction/255, h,50)
cv2.imshow('restoration image', deconvolved)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Add gausse noise
noise_mean = 0;
noise_var = 0.00001;
unction_gauss=util.random_noise(unction, mode='gaussian')
cv2.imshow('unction gauss image', unction_gauss)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Restoration by Lucy-Richardson
deconvolved = restoration.richardson_lucy(unction_gauss, h,18)
cv2.imshow('restoration image', deconvolved)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
