```python
import numpy as np
import cv2
```

```python
img = cv2.imread('images/Kianu.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img=cv2.resize(img,(400,400))
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fast = cv2.FastFeatureDetector_create(40)
fast.setNonmaxSuppression(False)

kp = fast.detect(img, None)
kp_image = cv2.drawKeypoints(img, kp, None, color=(0, 0, 255))

cv2.imshow('FAST', kp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('images/Kianu.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image = np.float32(gray_image)
dst = cv2.cornerHarris(gray_image, blockSize=2, ksize=3, k=0.04)
dst = cv2.dilate(dst, None)
image[dst > 0.01 * dst.max()] = [255, 0, 0]
cv2.imshow('Harris corner', cv2.resize(image, (400,400)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
orb = cv2.ORB_create(nfeatures=500)
kp, des = orb.detectAndCompute(img, None)

# Drawing the keypoints
kp_image = cv2.drawKeypoints(img, kp, None, color=(0, 255, 0), flags=0)

cv2.imshow('ORB', kp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
