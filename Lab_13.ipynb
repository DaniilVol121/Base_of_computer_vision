```python
import numpy as np
import cv2
import skimage
```

```python
img = cv2.imread('Images/Chess.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img,(450, 600))
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
retval, corners = cv2.findChessboardCorners(img, (8,5), cv2.CALIB_CB_ADAPTIVE_THRESH +cv2.CALIB_CB_FAST_CHECK + cv2.CALIB_CB_NORMALIZE_IMAGE)
fnl = cv2.drawChessboardCorners(img, (8, 5), corners, retval)
cv2.imshow('fnl', fnl)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
